```python
poem='''白日依山盡
黃河入海流
欲窮千里目
更上一層樓
'''

poem=poem.replace('\n','')

x=len(poem)
count=0
for i in range(1,x):
  if x < 0:
    break;
  else:
    x-=i
    count+=1

index=0

for row in range(1,count+1):
    for k in range(row):
        if index < len(poem):
          print(poem[index],end='')
          index+=1
    print()
```

```python
poem='''白日依山盡
黃河入海流
欲窮千里目
更上一層樓
'''

msg="<style> th { padding: 5px; border: solid 1px; } </style>"

poem=poem.replace('\n','')

x=len(poem)
count=0
for i in range(1,x):
  if x < 0:
    break;
  else:
    x-=i
    count+=1

index=0
msg+="<table>\n"
for row in range(1,count+1):
  msg+="<tr>\n"
  for k in range(row):
    if index < len(poem):
      msg+=f"<th>{poem[index]}</th>"
      index+=1
  msg+="\n"
  msg+="</tr>\n"
msg+="</table>\n"


fname = "see.html"
with open(fname, "w") as fp:
    fp.write(msg)
```
